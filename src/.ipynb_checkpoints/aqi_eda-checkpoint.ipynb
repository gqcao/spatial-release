```r
DATA_PATH = "../data/"
```

```r
#load(paste0(DATA_PATH, 'cal_aqi_pred.RData'))
trn_data <- read.csv(paste0(DATA_PATH, 'daily_88101_2016_cal_conc.csv'))
```

```r
names(trn_data)
dim(unique(trn_data['Local.Site.Name']))
```

```r
by_site_mean <- aggregate(trn_data[, 'AQI'], list(trn_data$City.Name), mean)
names(by_site_mean) <- c('Stations', 'AQI')
ggplot(by_site_mean, aes(x=reorder(Stations, AQI), y=AQI, fill=AQI)) + 
    geom_bar(stat="identity", width=.5) + 
    coord_flip() +
    scale_fill_gradient2(mid='green', high='orange', space='Lab') + 
    labs(title="Average AQI in California 2016", x='Cites')
```

```r
site_over_50 <- by_site_mean[by_site_mean$AQI > 50, ]
site_over_50
names(trn_data)
top5_cities <- unique(subset(trn_data, City.Name %in% site_over_50$Stations, select=c(City.Name, County.Name, Address, Local.Site.Name)))
top5_cities[!duplicated(top5_cities$City.Name),]
```

```r
by_site_mean[which(by_site_mean$AQI == max(by_site_mean$AQI)),]
```

```r
ontario_df <- subset(trn_data, City.Name=='Ontario', select=c('AQI', 'Date.Local'))
ontario_df$Date.Local <- as.Date(ontario_df$Date.Local)

ontario_by_month <- ontario_df %>%
    mutate(month = format(Date.Local, "%m")) %>%
    group_by(month) %>%
    summarise(avg.aqi = mean(AQI))
ggplot(ontario_by_month, aes(x=month, y=avg.aqi)) + 
    geom_bar(stat="identity", fill="orange") +    
    coord_cartesian(ylim=c(20, 65)) +
    labs(title="2016 AQI in Ontario CA by Month", x='Month', y='AQI')
```

```r
bakersfield_df <- subset(trn_data, City.Name=='Bakersfield', select=c('AQI', 'Date.Local'))
bakersfield_df$Date.Local <- as.Date(bakersfield_df$Date.Local)

bakersfield_by_month <- bakersfield_df %>%
    mutate(month = format(Date.Local, "%m")) %>%
    group_by(month) %>%
    summarise(avg.aqi = mean(AQI))
ggplot(bakersfield_by_month, aes(x=month, y=avg.aqi)) + 
    geom_bar(stat="identity", fill="orange") +    
    labs(title="2016 AQI in Bakersfield CA by Month", x='Month', y='AQI') +
    coord_cartesian(ylim=c(20, 85))    
```

```r
corcoran_df <- subset(trn_data, City.Name=='Corcoran', select=c('AQI', 'Date.Local'))
corcoran_df$Date.Local <- as.Date(corcoran_df$Date.Local)

corcoran_by_month <- corcoran_df %>%
    mutate(month = format(Date.Local, "%m")) %>%
    group_by(month) %>%
    summarise(avg.aqi = mean(AQI))
ggplot(corcoran_by_month, aes(x=month, y=avg.aqi)) + 
    geom_bar(stat="identity", fill="orange") +    
    labs(title="2016 AQI in Corcoran CA by Month", x='Month', y='AQI') +
    coord_cartesian(ylim=c(20, 80))    
```

#### CA spatial plot refers to  https://eriqande.github.io/rep-res-web/lectures/making-maps-with-R.html

```r
states <- map_data("state")
ca_df <- subset(states, region == "california")
counties <- map_data("county")
ca_county <- subset(counties, region == "california")
```

```r
ca_base <- ggplot(data = ca_df, mapping = aes(x = long, y = lat, group = group)) + 
  coord_fixed(1.3) + 
  geom_polygon(color = "black", fill = "gray")
```

```r
capitalize_first_letter <- function(x) {
  s <- strsplit(x, " ")[[1]]
  paste(toupper(substring(s, 1,1)), substring(s, 2),
      sep="", collapse=" ")
}
```

```r
by_county_mean <- aggregate(trn_data[, 'AQI'], list(trn_data$County.Name), mean)
names(by_county_mean) <- c('subregion','AQI')
min(by_county_mean$AQI)
max(by_county_mean$AQI)
ca_county$subregion <- sapply(ca_county$subregion, capitalize_first_letter)
```

```r
ca_county_aqi <- inner_join(ca_county, by_county_mean, by = "subregion")
ca_county_aqi <- data.frame(ca_county_aqi)
```

```r
ca_county_aqi_map <- ca_base + 
    geom_polygon(data = ca_county_aqi, aes(fill = AQI), color = "white") +
    geom_polygon(color = "black", fill = NA) +
    theme_bw() + scale_fill_gradient2(mid='green', high='orange', space='Lab') +
    labs(title="Average AQI in California 2016", x='Longitude', y='Latitude')
ca_county_aqi_map
```

```r
pred_df <- read.csv(paste0(DATA_PATH, 'cal_aqi_pred.csv'))
```

```r
min(trn_data$AQI)
max(trn_data$AQI)
min(pred_df$Pred_AQI)

#pred_df$Pred_AQI <- (pred_df$Pred_AQI - min(pred_df$Pred_AQI)) / (max(pred_df$Pred_AQI) - min(pred_df$Pred_AQI)) * 157
max(pred_df$Pred_AQI)
```

```r
by_county_median <- aggregate(pred_df[, 'Pred_AQI'], list(pred_df$county), median)
names(by_county_median) <- c('subregion','AQI')
by_county_median$AQI <- (by_county_median$AQI - min(by_county_median$AQI)) / (max(by_county_median$AQI) - min(by_county_median$AQI)) * 43 + 12
```

```r
by_county_median$subregion <- as.character(by_county_median$subregion)
by_county_median[which(by_county_median$subregion=='SF'),1] <- 'San Francisco'
by_county_median$subregion <- gsub(' County', '', by_county_median$subregion)
```

```r
ca_county_aqi <- inner_join(ca_county, by_county_median, by = "subregion")
ca_county_aqi <- data.frame(ca_county_aqi)
ca_county_aqi_map <- ca_base + 
    geom_polygon(data = ca_county_aqi, aes(fill = AQI), color = "white") +
    geom_polygon(color = "black", fill = NA) +
    theme_bw() + scale_fill_gradient2(mid='green', high='orange', space='Lab') +
    labs(title="Average AQI in California 2016", x='Longitude', y='Latitude')
ca_county_aqi_map
```

```r
la_df <- pred_df[which(pred_df$'county'=='Los Angeles'),]
dim(la_df)

breaks <- seq(0, 350, 50)
labels <- breaks[1:7]+1
la_df$level <- cut(la_df$Pred_AQI, breaks=breaks, labels=labels)
la_df$level <- as.numeric(as.character(la_df$level))
la_by_city <- la_df %>% group_by(city) %>% dplyr::count(level)
double_day_city <- la_df %>% group_by(city) %>% dplyr::count(city)
double_day_city <- double_day_city[which(double_day_city$n == 732),'city']
la_by_city[la_by_city$city %in% double_day_city$city, 'n'] <- round(la_by_city[la_by_city$city %in% double_day_city$city, 'n']/2)
names(la_by_city) <- c('City', 'Level', 'Days')
```

```r
#la_by_city_sorted
la_by_city_sorted <- la_by_city %>%  group_by(Level) %>% arrange(Days)
la_by_city_sorted <- la_by_city_sorted[la_by_city_sorted$Level=='1','City']
top50_la_city <- la_by_city[la_by_city$City %in% la_by_city_sorted$City[1:50],]
top50_la_city
```

```r
# Stacked + percent
top50_la_city %>% group_by(Level) %>%
    mutate(City = fct_reorder(City, -Days)) %>%    
    ggplot(aes(fill=Level, y=Days, x=City)) + 
    #geom_bar(position="fill", stat="identity", width=.5) +
    geom_bar(stat="identity", width=.5) +
    scale_fill_gradient2(mid='green', high='red', space='Lab') +
    labs(title="50 Most Polluted Cities in Los Angeles County 2016") +    
    theme(text = element_text(size=7)) +
    coord_flip()
```
